```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python

num_obs = 50
x1 = np.random.uniform(low=-10.0, high=10.0, size=num_obs)
noise = np.random.normal(loc=0.0, scale=5.0, size=num_obs)
y = 10 + 2 * x1 + noise

# Create a DataFrame
data = pd.DataFrame({'x1': x1, 'y': y})

# Fit the linear regression model
lr = LinearRegression().fit(data[['x1']], data['y'])

# Plot the original data and the regression line
plt.figure(figsize=(10, 7))
plt.scatter(data['x1'], data['y'], color='blue', label='Original data')
plt.plot(data['x1'], lr.predict(data[['x1']]), color='red', label='Regression line')
plt.xlabel('x1')
plt.ylabel('y')
plt.title('Scatter plot of x1 vs y with regression line')
plt.legend()
plt.show()

# Predict new values
x_new = np.array([0, 1, 2, 3]).reshape(-1, 1)
y_prediction = lr.predict(x_new)
print("Predicted values for x_new:", y_prediction)
```
