```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return x**2 + y**2 + 4

def grad(x, y):
    df_dx = np.cos(y) + 0.2 * x + 1
    df_dy = - np.sin(y) + 0.2 * y - 1
    return np.array([df_dx, df_dy])

def grad_desc(init_point, lr, iters):
    point = np.array(init_point, dtype=float)
    history = []

    for _ in range(iters):
        gradient = grad(point[0], point[1])
        point = point - lr * gradient
        history.append(point.copy())

    return point, history

init_point = [10, 10]
lr = 0.1
iters = 100

min_point, history = grad_desc(init_point, lr, iters)

print("The minimum point is:", min_point)
print("Minimum value of the function:", f(min_point[0], min_point[1]))

# Plot the history of the points
history = np.array(history)
plt.plot(history[:, 0], history[:, 1], 'o-', markersize=3, label='Path of gradient descent')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gradient Descent Path')
plt.legend()
plt.show()
```
