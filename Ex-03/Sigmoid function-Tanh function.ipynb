```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random values
vals = np.random.randn(100)

# Compute sigmoid values
sigmoid_vals = 1 / (1 + np.exp(-vals))

# Compute tanh values
tanh_vals = np.tanh(vals)

# Plot the results
plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)
plt.scatter(vals, sigmoid_vals, color='blue', label='Sigmoid', marker='.')
plt.title('Sigmoid')
plt.subplot(1, 2, 2)
plt.scatter(vals, tanh_vals, color='red', label='Tanh', marker='.')
plt.title('Tanh')
plt.tight_layout()
plt.show()
```
